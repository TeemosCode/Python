```python
s =0
N = 1000000
for i in range(N):
    a, b = random(), random()
    if a**2 + b**2 <=1:
        s += 1
print s*4./N
```

```python
def error(N):
    s =0    
    for i in range(N):
        a, b = random(), random()
        if a**2 + b**2 <=1:
            s += 1
    return abs(s*4./N - pi)
```

```python
def avg_error(N):
    s = 0
    for i in range(1000):
        s+=error(N)
    return s/100.
```

```python
plot([1/avg_error(N)**2 for N in range(10, 1000, 10)])
```
